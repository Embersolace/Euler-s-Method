```python
"""
====================================================
 Introduction to Euler's Method
 Solving the ODE: y' = -2y,   y(0) = 1
 Exact solution: y(t) = e^(-2t)
====================================================
"""

import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# Define the differential equation y' = f(t, y)
# ----------------------------------------------------
def f(t, y):
    return -2 * y

# ----------------------------------------------------
# Exact solution for comparison
# ----------------------------------------------------
def exact_solution(t):
    return np.exp(-2 * t)

# ----------------------------------------------------
# Euler's Method
# ----------------------------------------------------
def euler_method(f, t0, y0, h, N):
    """
    Perform N steps of Euler's method.

    f  : function defining y' = f(t, y)
    t0 : initial time
    y0 : initial value y(t0)
    h  : step size
    N  : number of steps
    """
    t = np.zeros(N + 1)
    y = np.zeros(N + 1)

    t[0] = t0
    y[0] = y0

    for n in range(N):
        t[n + 1] = t[n] + h
        y[n + 1] = y[n] + h * f(t[n], y[n])  # Euler update

    return t, y

# ----------------------------------------------------
# Parameters
# ----------------------------------------------------
t0 = 0
y0 = 1
T = 2            # final time
h = 0.2     # step size
N = int((T - t0) / h)

# ----------------------------------------------------
# Run Euler's method
# ----------------------------------------------------
t_vals, y_euler = euler_method(f, t0, y0, h, N)

# Exact curve for comparison
t_exact = np.linspace(0, T, 300)
y_exact = exact_solution(t_exact)

# ----------------------------------------------------
# Compute error at t = 2
# ----------------------------------------------------
error = abs(y_exact[-1] - y_euler[-1])
rel_error= error/abs(y_exact[-1])
print("T=", T)
print("Euler approximation at t = T:", y_euler[-1])
print("Exact value at t = T:", y_exact[-1])
print("Absolute error:", error)
print("Percent Relative error:", rel_error*100)

# ----------------------------------------------------
# Plot the solution
# ----------------------------------------------------
plt.figure(figsize=(9, 5))
plt.plot(t_exact, y_exact, 'k-', label="Exact Solution")
plt.plot(t_vals, y_euler, 'ro--', label="Euler's Method")

plt.title("Introduction to Euler's Method")
plt.xlabel("t")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

print(y_euler)
```
